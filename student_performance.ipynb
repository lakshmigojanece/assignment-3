```python
#Study hours vs Performance
df.groupby('Hours Studied')['Performance Index'].mean()
```

```python
plt.figure()
sns.scatterplot(x="Hours Studied", y="Performance Index", data=df)
plt.title("Study Hours vs Performance")
plt.show()
```

```python
#insights----> low hours studied- low performance --->affect admission--->affect revenue loss
            #  high hours studied- high performance ---->high admission -----> revenue is high profit
            #  studied hours  one by one increase performance also increase.....
```

```python
#Sleep vs Performance
df.groupby('Sleep Hours')['Performance Index'].mean()
```

```python
plt.figure()
sns.boxplot(x="Sleep Hours", y="Performance Index", data=df)
plt.title("Sleep Hours vs Performance")
plt.show()

```

```python
#insights------> sleep hours decrease performance is very low
                # sleep hours increase performance also increase
```

```python
#Extracurricular activities vs Performance
df.groupby('Extracurricular Activities')['Performance Index'].mean()
```

```python
sns.boxplot(x="Extracurricular Activities",
            y="Performance Index", data=df)
plt.title("Extracurricular Activities Impact")
plt.show()
```

```python
#insights -------> extracurricular activities students performance is also good
                  # extracurricular not activities students performance is not good
```

```python
#Previous Scores vs Performance

df.groupby('Previous Scores')['Performance Index'].mean()
```

```python
plt.figure()
sns.scatterplot(x="Previous Scores", y="Performance Index", data=df)
plt.title("Previous Scores Impact")
plt.show()
```

```python
#insights ------> students performance good scores in pervious and also increase performance score
                 # students is not performance scores in pervious and also decrease performance score
```

```python
#Sample Question Papers Practiced vs Performance

df.groupby('Sample Question Papers Practiced')['Performance Index'].mean()
```

```python
plt.figure()
sns.boxplot(x="Sample Question Papers Practiced",y="Performance Index", data=df)
plt.title("Sample Question Papers Practiced vs Performance")
plt.show()
```

```python
# insights -----> students practice in sample question paper proper and also perfermance is increase
                # students practice in sample question paper not proper and also perfermance is not increase
```
